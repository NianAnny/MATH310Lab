```python
import numpy as np  
import matplotlib.pyplot as plt  
```

```python
def basin_newton(f, df, initial_guess, tol): # for real-valued functions, well-performed when roots are integers or rational numbers
    initial = initial_guess
    
    final_approximation = [] # final root found, avoiding repetition
    
    root = [] # roots found for each initial guess
    approx = [[] for _ in range(len(initial))] # points that indicates how to approximate to the root at each iteration
    
    for i in range(len(initial)):
        if df(initial[i]) == 0 or np.abs(df(initial[i]))<tol:
            raise Exception(f"The first derivative is zero at the point of x = {initial[i]} (position {i} in the given list), so cannot execute Newton's Method.")
        
        approx[i].append(initial[i]) # add the start point to each iteration
        
        if np.abs(f(initial[i])) < tol:
            root.append(initial[i]) # check if the start point is already a root
        
        while np.abs(f(initial[i])) > tol:
            f_x = f(initial[i])
            df_x = df(initial[i])
            
            initial[i] = initial[i] - f_x/df_x # newton's method formula for root-finding 
            
            approx[i].append(initial[i]) # approximations for each iteration
            
            if np.abs(f(initial[i])) < tol:
                root.append(initial[i]) # add ended approximations (root) for each initial, may have the same root.


        # final answers with no repetitions
        final_approximation = list(set(root)) 
        
        
    return final_approximation, root, approx
```

<h1>1. </h1> 

$f(x) = (x-4)(x+1)$ 

```python
def f(x):
    return x**2 - 3*x - 4

def df(x):
    return 2*x - 3


#initial_guess = [5, 4, -2, 1, 2, -3, 6, -1, 3]
initial_guess = np.arange(-5, 6, 0.1)

# Outputs for Basins of Attraction
final_approxs, roots, approxs = basin_newton(f, df, initial_guess, 1e-16)

print(f"Final approximations: {final_approxs}")
#print(f"Approximations for each initial guess: {roots}")
#print(f"Iterations for each initial guess: {approxs}")


colors = ["red", "blue", "green"]

# figure plot
plt.figure(figsize = (9,7))
x= np.linspace(-2, 5, 400)
plt.plot(x, f(x))

for i in range(len(roots)):
    if np.abs(approxs[i][-1] - final_approxs[0]) <= 1e-10:
        for j in range(len(approxs[i])):
            plt.plot(approxs[i][0], 0, ".", color = colors[0])
            
    else:
        for j in range(len(approxs[i])):
            plt.plot(approxs[i][0], 0, ".", color = colors[1])
            
plt.plot(final_approxs[0], 0,"o", ms = 10, color = colors[0], label = f"root 1 at x = {final_approxs[0]}")
plt.plot(final_approxs[1], 0, "o", ms = 10, color = colors[1], label = f"root 2 at x = {final_approxs[1]}")

plt.legend()              
plt.xlim(-3, 6)  
plt.title("Basins of Attraction for $f(x) = (x-4)(x+1)$")
plt.show()          
```

</h1>

$f(z) = z^3 - 1$, where $z$ is a complex variable

```python
def newton(f, df, x_0, tol):
    x = x_0
    
    while np.abs(f(x)) > tol:
        f_x = f(x)
        df_x = df(x)
        
        if df_x == 0 or np.abs(df_x)<tol:
            raise Exception("The second derivative is zero at the point, so cannot execute Newton's Method.")
            
        x_new = x - f_x/df_x
        
        x = x_new
        
    return x


def f_complex(z):
    return z**3 - 1

def df_complex(z):
    return 3*z**2


# Set up the grid of complex numbers
x = np.linspace(-2, 2, 400) # for real numbers
y = np.linspace(-2, 2, 400) # for imaginary numbers
X, Y = np.meshgrid(x, y)
Z = X + 1j*Y # complex numbers


# Initialize an array to store the results
roots = np.zeros(Z.shape, dtype=complex)

for i in range(len(x)): # real numbers of z
    for j in range(len(y)): # imaginary numbers of z
        root = newton(f_complex, df_complex, Z[i, j], 1e-6) # apply Newton's method to each point in the grid
        if root is None:
            roots[i, j] = np.nan  # Assign NaN to points where the method fails to converge

        else:
            roots[i, j] = root
            


# Known roots of z^3 - 1
known_roots = [1, (-1 + np.sqrt(3) * 1j)/2, (-1 - np.sqrt(3) * 1j)/2]


# Map each point to the index of its closest root in the known_roots array
root_indices = np.zeros(Z.shape, dtype=int)

for idx, root in enumerate(known_roots): # add index of the root to each iteration 
    check = np.isclose(roots, root, atol=1e-6) # check if the element in roots is close to that in know_roots array, absolute tolerance of 1e-6
    root_indices[check] = idx

# Plot
plt.figure(figsize = (9,7))
c = plt.pcolormesh(X, Y, root_indices, cmap = 'rainbow')
plt.colorbar(c)
plt.title('Basins of Attraction for $f(z) = z^3 - 1$')
plt.xlabel('Real(z)')
plt.ylabel('Imaginary(z)')
plt.show()
```
